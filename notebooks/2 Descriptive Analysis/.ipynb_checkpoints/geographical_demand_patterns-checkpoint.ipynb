# Geographical Demand Patterns

In this notebook we analyse the popularity of the sharing stations in Chicago. We describe the patterns that we observed and provide an explanation.

**Hint**: This notebook needs to executed in order to view the interactive plots. Furthermore, please trust the notebook (in the top right corner) in order to see the maps.

## 1 Imports

```python
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

import calendar
from datetime import datetime, timedelta
import warnings
warnings.simplefilter(action='ignore', category=UserWarning)
```

Load Bike Dataset

```python
df_chicago = pd.read_csv('../../dataset/chicago_2018_clean.csv') 
df_chicago.head()
```

```python
#Format the columns from datatype object to datatype datetime
df_chicago['start_time'] = pd.to_datetime(df_chicago['start_time'])
df_chicago['end_time'] = pd.to_datetime(df_chicago['end_time'])
```

## 2 Overall Popularity

### 2.1 Most popular stations throughout the whole year

First we consider the most popular bike stations throughout the whole year. We measure popularity by counting the amount of bike rides which involve the particular station. Meaning, we count every bike ride which either starts or ends at the particular station.

```python
start_counts = df_chicago.groupby("start_station_name").size()
end_counts = df_chicago.groupby("end_station_name").size()
total_counts = start_counts.add(end_counts, fill_value = 0).sort_values(ascending=False)
```

```python
# Select the top stations and their total ride counts
top_stations = total_counts.head(10)

# Extract the station names and ride counts into separate arrays
top_station_names = top_stations.index
ride_counts = top_stations.values

# Set the bar width
bar_width = 0.4

# Set the bar positions
bar_positions = range(len(top_stations))

# Create the bar plot
plt.bar(bar_positions, ride_counts, width=bar_width)

# Set the x-axis tick labels
plt.xticks(bar_positions, top_station_names, rotation=90)

# Set the plot title and axes labels
plt.title("Top Bike Sharing Stations")
plt.xlabel("Station")
plt.ylabel("Number of Rides")

# Show the plot
plt.show()
```

The above plotted stations are the 20 most popular stations throughout the whole year. Next we look at the location of these stations in the city of chicago.

#### Import station dataset

```python
df_stations = pd.read_csv('../../dataset/chicago_stations.csv')
df_stations.head()
```

#### Create station popularity dataframe

```python
df_stations_indexed_by_id = df_stations.set_index('ID')
```

```python
start_counts_by_id = df_chicago.groupby("start_station_id").size()
end_counts_by_id = df_chicago.groupby("end_station_id").size()
total_counts_by_id = start_counts_by_id.add(end_counts_by_id, fill_value = 0).sort_values(ascending=False)
```

```python
df_station_popularity = pd.DataFrame(total_counts_by_id, columns=['popularity'])
df_station_popularity = df_station_popularity.reset_index().rename(columns={'index': 'ID'})
df_station_popularity = df_station_popularity.sort_values(by = 'ID')
df_station_popularity['position'] = df_station_popularity.apply(
    lambda row: df_stations_indexed_by_id.loc[row['ID']].position,
    axis = 1
)
df_station_popularity = df_station_popularity.sort_values(by = 'popularity', ascending = False)
```

#### Location of most popular bike stations

```python
import folium
from ast import literal_eval as make_tuple

def create_map(coordinates, zoom = 11, radius = 75, color = 'crimson', offset = (0, 0)):
    origin = coordinates.iloc[0]
    m = folium.Map(location=[make_tuple(origin)[0] + offset[0], make_tuple(origin)[1] + offset[1]], zoom_start=zoom)

    for coord in coordinates:
        folium.Circle(
            radius=radius,
            location=make_tuple(coord),
            color=color,
            fill=False,
        ).add_to(m)

    return m
```

```python
create_map(df_station_popularity.position.dropna().head(20), zoom = 13, radius = 100, offset = (0.01, 0))
```

The map shows that most of the popular stations seem to be located near the center of the city. There are a few outliers, most noteably the stations 'Theater on the Lake' and 'Lake Shore Dr & North Blvd', though even these stations are fairly close to the center of the city. The popularity of those stations will be analysed later. In the following we will analyse whether there is a correlation between the popularity of a station and its distance from the center of the city across the entire dataset.

As the center we choose Michigan Ave & Washington St, as it is not the geographical center of the city but the most high traffic street in chicago.

```python
center = df_stations.loc[
    df_stations['station_name'] == 'Michigan Ave & Washington St'
]

center_x = center.x.iloc[0]
center_y = center.y.iloc[0]
```

```python
import math

def dist_from_center(x, y):
    return math.dist([center_x, center_y], [x, y])
```

```python
df_station_popularity['distance_from_center'] = df_station_popularity.apply(
    lambda row: dist_from_center(
        df_stations_indexed_by_id.loc[row['ID']].x,
        df_stations_indexed_by_id.loc[row['ID']].y
    ) * 100,
    axis = 1
)
df_station_popularity = df_station_popularity.sort_values(by = 'popularity', ascending = False)
```

```python
%matplotlib inline

plt.scatter(df_station_popularity.distance_from_center, df_station_popularity.popularity, sizes=[6])
plt.title("Correlation between popularity and distance from center")
plt.xlabel("Distance from center of the city in miles")
plt.ylabel("Number of Rides")
plt.show()
```

As the plot shows, the popularity of a station seems to decrease when its distance from the center increases. Stations that are more than 10 miles away from the center barely reach an involvement of 10000 rides, while stations less than 5 miles away from the center are commonly involved in more than 20000 rides. Furthermore, this plot again underlines how the most popular stations are fairly close to the center of the city. Interestingly enough, there are also few stations which are very close to the center but are rarely used.

```python
central_but_unpopular_station = df_station_popularity.sort_values(
    by = ['distance_from_center'], ascending = True
).head(100).sort_values(
    by = 'popularity', ascending = True
).head(5)
central_but_unpopular_station['station_name'] = central_but_unpopular_station.apply(
    lambda row: df_stations_indexed_by_id.loc[row.ID]['station_name'],
    axis = 1
)
central_but_unpopular_station
```

These are the most unpopular stations among the top 100 closest stations to the center. When looking at the first ride which was taken from one of these stations one can see that all three of these stations were introduced in mid september, thus explaining the low amount of rides involved in these stations.

## 4 Temporal Popularity

How popular a particular station is, can change over time. There might be stations which are especially popular in the summer months while being less popular in the winter months, or stations which are busy during the night and less during the day. For this reason we considered the temporal popularity. The following section will present the popularity we observed when looking at the monthly and daily popularity of stations.

### 4.1 Monthly Popularity

```python
def get_counts_for_month(month):
    start_counts = df_chicago.loc[
        (df_chicago['start_time'].dt.month == month) | 
        (df_chicago['end_time'].dt.month == month)
    ].groupby("start_station_name").size()
    end_counts = df_chicago.loc[
        (df_chicago['start_time'].dt.month == month) | 
        (df_chicago['end_time'].dt.month == month)
    ].groupby("end_station_name").size()
    return start_counts.add(end_counts, fill_value = 0).sort_values(ascending=False)

def get_start_counts_for_month(month):
    return df_chicago.loc[
        (df_chicago['start_time'].dt.month == month) | 
        (df_chicago['end_time'].dt.month == month)
    ].groupby("start_station_name").size().sort_values(ascending=False)

def get_end_counts_for_month(month):
    return df_chicago.loc[
        (df_chicago['start_time'].dt.month == month) | 
        (df_chicago['end_time'].dt.month == month)
    ].groupby("end_station_name").size().sort_values(ascending=False)
```

```python
top_stations_by_month = {}
top_start_stations_by_month = {}
top_end_stations_by_month = {}

for month in range(1, 13):
    top_start_stations_by_month[month] = get_start_counts_for_month(month)
    top_end_stations_by_month[month] = get_end_counts_for_month(month)
```

```python
for month in range(1, 13):
    top_stations_by_month[month] = pd.DataFrame(
        {
            'start': top_start_stations_by_month[month], 
            'end': top_end_stations_by_month[month],
        }
    ).fillna(0)
    top_stations_by_month[month]['sum'] = top_stations_by_month[month]['start'] + top_stations_by_month[month]['end']
    top_stations_by_month[month] = top_stations_by_month[month].sort_values(
        by = 'sum', ascending = True
    )
```

```python
%matplotlib notebook

init_top_stations = top_stations_by_month[1].tail(20)
start_data = init_top_stations['start']
end_data = init_top_stations['end']

fig, ax = plt.subplots(figsize=(8, 5))
labels = list(init_top_stations.index)

bars_start = ax.barh(labels, start_data, label='start')
bars_end = ax.barh(labels, end_data, left=start_data, label='end')

ax.set_title("Top Bike Sharing Stations By Month")
ax.set_xlabel('Number of Rides ')
ax.set_ylabel("Stations")

ax.legend()

fig.subplots_adjust(left=0.25, bottom=0.35)

axfreq = fig.add_axes([0.25, 0.01, 0.65, 0.03])
month_slider = Slider(
    ax=axfreq,
    label='Month',
    valmin=1,
    valmax=12,
    valinit=1,
    valstep=range(1, 13)
)

def update(val):
    top_stations = top_stations_by_month[val].tail(20)
    all_bars = []
    for bar, h in zip(bars_start, top_stations['start'].values):
        bar.set_x(0)
        bar.set_width(h)
    for bar, h_start, h_end in zip(bars_end, top_stations['start'].values, top_stations['end'].values):
        bar.set_x(h_start)
        bar.set_width(h_end)
    ax.set_yticks(range(len(top_stations)) , top_stations.index)
    ax.set_xlim([0, top_stations.max().max() * 1.1])
    ax.autoscale_view()
    fig.canvas.draw_idle()
    
month_slider.on_changed(update)

plt.tight_layout()
plt.show()
```

When looking at the data for January (1), the first thing that is striking is that the station Streeter Dr & Grand Ave, which we previously determined to be the overall most popular station is not even listed in the top 20. In fact it is only the 498th most popular station in january, with only 958 rides involving the station.

Since the station is located right next to the Milton Lee Olive and Plk Bros Park and is the station with the closest access to the ohio street beach, its popularity is most likely proportional to the popularity of these destinations. ### Import weather dataset

```python
df_chicago_weather = pd.read_csv('../../dataset/chicago_weather_2018.csv')
df_chicago_weather['date_time'] = pd.to_datetime(df_chicago_weather['date_time'])
df_chicago_weather.head()
```

### 4.2 Correlation between weather and station popularity

```python
%matplotlib inline

streeter_station = {}
average_weather = {}

for month in range(1, 13):
    try:
        streeter_station[month] = top_stations_by_month[month].loc[
            'Streeter Dr & Grand Ave'
        ]['sum']
    except KeyError:
        streeter_station[month] = 0
    average_weather[month] = df_chicago_weather.loc[
        df_chicago_weather['date_time'].dt.month == month
    ]['max_temp'].mean()

fig, ax = plt.subplots()
ax.plot(range(1, 13), streeter_station.values(), color='blue')
ax.set_ylabel("Number of Rides", color='blue')

ax2=ax.twinx()
ax2.plot(range(1, 13), average_weather.values(), color='red')
ax2.set_ylabel("Highest Temperature", color='red')

plt.title("Popularity of Streeter Dr & Grand Ave and Highest Temperature per month")
plt.xlabel("Month")
plt.show()
```

The above plot shows the number of rides recorded for the station Streeter Dr & Grand Ave in each month and the highest temperature recorded for that month. As one can see a higher temperature seems to be correlated with a higher amount of rides. This could be an explanation for why Streeter Dr & Grand Ave is very unpopular in the winter months like december and january and spikes in popularity during the summer. It appears that at a temperature of around 24 degrees celsius the maximum of people are attracted to Lake Michigan and the ohio street beach near Streeter Dr & Grand Ave.

```python
%matplotlib inline

lake1 = {}
lake2 = {}
millennium_park = {}

for month in range(1, 13):
    try:
        lake1[month] = top_stations_by_month[month].loc[
            'Lake Shore Dr & North Blvd'
        ]['sum']
    except KeyError:
        lake1[month] = 0
    try:
        lake2[month] = top_stations_by_month[month].loc[
            'Lake Shore Dr & Monroe St'
        ]['sum']
    except KeyError:
        lake2[month] = 0
    try:
        millennium_park[month] = top_stations_by_month[month].loc[
            'Millennium Park'
        ]['sum']
    except KeyError:
        millennium_park[month] = 0

fig, ax = plt.subplots()
ax.plot(range(1, 13), lake1.values(), color='blue', label='Lake Shore Dr & North Blvd')
ax.plot(range(1, 13), lake2.values(), color='green', label='Lake Shore Dr & Monroe St')
ax.plot(range(1, 13), millennium_park.values(), color='yellow', label='Millennium Park')
ax.set_ylabel("Number of Rides", color='blue')
ax.legend()

ax2=ax.twinx()
ax2.plot(range(1, 13), average_weather.values(), color='red', label='Temperature')
ax2.set_ylabel("Highest Temperature", color='red')
ax2.legend()

plt.title("Popularity of Streeter Dr & Grand Ave and Highest Temperature per month")
plt.xlabel("Month")
plt.show()
```

Similarly, the popularity of the stations Lake Shore Dr & Monroe St, Lake Shore Dr & North Blvd and Millennium Park also seems to be correlated with the temperature. Since all three of these stations are either located near Lake Michigan or a popular park, the same reasoning for their popularity can be applied as with the station Streeter Dr & Grand Ave.

It is important to note, that a similar pattern can be observed with any of the top stations. A higher temperature seems to be generally correlated to a higher amount of rides. However, the four stations which were analysed above had the unique pattern (among the top stations) of being particulary popular during summer months and almost never used during winter months. Whether the cause of this pattern truly is the seasonal change of the temperature cannot be said with certainty. However, it appears to be a fitting and also logical explanation.

## 4.2 Popularity By Hour

```python
def get_counts_for_hour(hour):
    start_counts = df_chicago.loc[
        (df_chicago['start_time'].dt.hour == hour) | (df_chicago['end_time'].dt.hour == hour)
    ].groupby("start_station_name").size()
    end_counts = df_chicago.loc[
        (df_chicago['start_time'].dt.hour == hour) | (df_chicago['end_time'].dt.hour == hour)
    ].groupby("end_station_name").size()
    return start_counts.add(end_counts, fill_value = 0).sort_values(ascending=False)

def get_start_counts_for_hour(hour):
    return df_chicago.loc[
        (df_chicago['start_time'].dt.hour == hour) | (df_chicago['end_time'].dt.hour == hour)
    ].groupby("start_station_name").size().sort_values(ascending=False)

def get_end_counts_for_hour(hour):
    return df_chicago.loc[
        (df_chicago['start_time'].dt.hour == hour) | (df_chicago['end_time'].dt.hour == hour)
    ].groupby("end_station_name").size().sort_values(ascending=False)
```

```python
top_stations_by_hour = {}
top_start_stations_by_hour = {}
top_end_stations_by_hour = {}

for hour in range(0, 24):
    top_start_stations_by_hour[hour] = get_start_counts_for_hour(hour)
    top_end_stations_by_hour[hour] = get_end_counts_for_hour(hour)
```

```python
for hour in range(0, 24):
    top_stations_by_hour[hour] = pd.DataFrame(
        {
            'start': top_start_stations_by_hour[hour], 
            'end': top_end_stations_by_hour[hour],
        }
        
    ).fillna(0)
    top_stations_by_hour[hour]['sum'] = top_stations_by_hour[hour]['start'] + top_stations_by_hour[hour]['end']
    top_stations_by_hour[hour] = top_stations_by_hour[hour].sort_values(by = 'sum', ascending = True).tail(20)
```

```python
%matplotlib notebook

init_top_stations = top_stations_by_hour[0]
start_data = init_top_stations['start']
end_data = init_top_stations['end']

fig, ax = plt.subplots(figsize=(8, 5))
labels = list(init_top_stations.index)

bars_start = ax.barh(labels, start_data, label='start')
bars_end = ax.barh(labels, end_data, left=start_data, label='end')

ax.set_title("Top Bike Sharing Stations By Hour")
ax.set_xlabel('Number of Rides ')
ax.set_ylabel("Stations")

ax.legend()

fig.subplots_adjust(left=0.25, bottom=0.35)

axfreq = fig.add_axes([0.25, 0.01, 0.65, 0.03])
month_slider = Slider(
    ax=axfreq,
    label='Hour',
    valmin=0,
    valmax=23,
    valinit=0,
    valstep=range(0, 24)
)

def update(val):
    top_stations = top_stations_by_hour[val]
    all_bars = []
    for bar, h in zip(bars_start, top_stations['start'].values):
        bar.set_x(0)
        bar.set_width(h)
    for bar, h_start, h_end in zip(bars_end, top_stations['start'].values, top_stations['end'].values):
        bar.set_x(h_start)
        bar.set_width(h_end)
    ax.set_yticks(range(len(top_stations)) , top_stations.index)
    ax.set_xlim([0, top_stations.max().max() * 1.1])
    ax.autoscale_view()
    fig.canvas.draw_idle()
    
month_slider.on_changed(update)

plt.tight_layout()
plt.show()
```

One interesting pattern that can be observed in the above plot is that most of the popular stations between 3am and 5am are used as end stations. This is probably a time range where a lot of people use the bikes to get home. As a consequence many of the popular station during this time are located further away from the center, where most people live.

```python
create_map(
    df_stations.set_index('station_name').loc[get_counts_for_hour(3).head(20).index].position, 
    zoom = 12, radius = 200, offset=(0.05, 0)
)
```

The map shows the 20 most popular stations at 3am. While most are still located near the center of the city, a considerable amount is shifted towards the north side of the city.

Considering Canal St & Adams St, it is interesting to observe that, most of the rides taken to or from that station occur between 3pm and 6pm. In contrast to the monthly popularity, the station does not consistently stay in the top 20 of stations.
