# KPI: Utilization of user types Customer and Subscriber 

This KPI provides information on the hourly, monthly, weekday, and general average usage of user types.

## 1 Import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta, date, time
import pickle
from matplotlib.ticker import MaxNLocator
```

Load Bike Dataset

```python
df_chicago =pd.read_csv('../dataset/chicago_2018_clean.csv') 

df_chicago
```

```python
#Convert into datetime objects
df_chicago['Time'] = pd.to_datetime(df_chicago['start_time']).dt.time
df_chicago["Hour"] = pd.to_datetime(df_chicago["start_time"]).dt.hour
df_chicago["Month"]= pd.to_datetime(df_chicago["start_time"]).dt.month
df_chicago["Weekday"]= pd.to_datetime(df_chicago["start_time"]).dt.dayofweek
df_chicago['Day'] = pd.to_datetime(df_chicago['start_time']).dt.date
df_chicago["Day"]= pd.to_datetime(df_chicago["Day"])
```

## 2 Percentage distribution of user types per hour

```python
# Create column for the hour of the trip
df_chicago['hour'] = pd.to_datetime(df_chicago['start_time']).dt.hour

# Count number of trips per hour and per user type
counts = df_chicago.groupby(['hour', 'user_type']).size().reset_index(name='counts')

# Create table with number of trips per hour and per user type
counts_table = counts.pivot(index='hour', columns='user_type', values='counts')

# Calculate percentage distribution of user types per hour
counts_table['Total'] = counts_table.sum(axis=1)
counts_table['Subscriber_percent'] = counts_table['Subscriber'] / counts_table['Total'] * 100
counts_table['Customer_percent'] = counts_table['Customer'] / counts_table['Total'] * 100

print(counts_table)
```

The table shows the hourly usage of customers and subscribers and indicates which user type accounts for what percentage of usage. 

```python
# Reorder the indexes
counts_table = counts_table.reindex(range(0,24))

# Create bar chart
counts_table[['Subscriber_percent', 'Customer_percent']].plot(kind='bar', stacked=True, title='Percentage distribution of user types per hour', grid=True, figsize=(10,5), color=['blue','black'])
plt.xlabel('Hour')
plt.ylabel('Percentage in %')
plt.legend(['Subscriber','Customer'],title="User Type",loc='lower right')
plt.show()
```

The bar chart shows that subscriber usage is higher than customer usage at all times of the day. It is also noticeable that subscriber usage is over 90% during the 5:00 - 8:00 a.m. time period. For customers, usage is highest around lunchtime between 10:00 am - 3:00 pm at about 26%. From this we can conclude that many subscribers cycle to work in the morning and return home after work.  

## 3 Percentage distribution of user types per month

```python
# Create column for the month of the trip
df_chicago['start_time'] = pd.to_datetime(df_chicago['start_time'])
df_chicago['month'] = df_chicago['start_time'].dt.month

# Count number of trips per month and per user type
counts = df_chicago.groupby(['month', 'user_type']).size().reset_index(name='counts')

# Create table with number of trips per month and per user type
counts_table = counts.pivot(index='month', columns='user_type', values='counts')

# Calculate percentage distribution of user types per month
counts_table['Total'] = counts_table.sum(axis=1)
counts_table['Subscriber_percent'] = counts_table['Subscriber'] / counts_table['Total'] * 100
counts_table['Customer_percent'] = counts_table['Customer'] / counts_table['Total'] * 100

print(counts_table)
```

The table shows the monthly usage of customers and subscribers and indicates which user type accounts for what percentage of usage.

```python
# Create bar chart
counts_table[['Subscriber_percent', 'Customer_percent']].plot(kind='bar', stacked=True, title='Percentage distribution of user types per month', grid=True, figsize=(10,5), color=['blue','black'])
plt.xlabel('Month')
plt.ylabel('Percentage in %')
plt.legend(['Subscriber','Customer'],title="User Type",loc='lower right')
plt.show()
```

The bar chart shows that customers use about 5% of the bikes in the winter months of January, February, March, October, November and December. However, this changes in the summer months and the percentage increases to about 25%. It can be concluded that subscribers also use the bike in winter because they pay the subscription every month, while customers in summer tend to use the service because they are very flexible and use the bike only when it suits them. 

## 4 Percentage distribution of user types per weekdays

```python
# Count number of trips per weekday and per user type
counts = df_chicago.groupby(['Weekday', 'user_type']).size().reset_index(name='counts')

# Create table with number of trips per weekday and per user type
counts_table = counts.pivot(index='Weekday', columns='user_type', values='counts')

# Calculate percentage distribution of user types per weekday
counts_table['Total'] = counts_table.sum(axis=1)
counts_table['Subscriber_percent'] = counts_table['Subscriber'] / counts_table['Total'] * 100
counts_table['Customer_percent'] = counts_table['Customer'] / counts_table['Total'] * 100

print(counts_table)
```

The table shows the weekday usage by customers and subscribers and indicates which user type accounts for what percentage of usage.

```python
# Create bar chart
counts_table[['Subscriber_percent', 'Customer_percent']].plot(kind='bar', stacked=True, title='Percentage distribution of user types per weekdays', grid=True, figsize=(10,5), color=['blue','black'])
plt.xlabel('Weekday')
plt.ylabel('Percentage in %')
plt.legend(['Subscriber','Customer'],title="User Type",loc='lower right')
plt.show()
```

The bar chart shows the usage on the individual weekdays. It is noticeable that customer usage increases from about 15% on weekdays to about 35% on weekends. This is an increase of 20%. It can be concluded that customers use the bicycles more often on weekends because they have more free time and therefore use the bicycle for spontaneous trips.

## 5 Annual average percentage of use of user types

```python
# Group by user type
grouped = df_chicago.groupby('user_type')

# Calculate total number of trips per user type
total_trips = grouped['user_type'].count()

# Calculate percentage of trips per user type
trips_percent = (total_trips / total_trips.sum()) * 100

print(trips_percent)
```

The percentages for the average annual usage of the individual user types show that 19% of customers and about 81% of subscribers use the bicycles annually.

```python
# Create diagram
trips_percent.plot(kind='bar')
colors = {'Customer': 'black', 'Subscriber': 'blue'}
trips_percent.plot(kind='bar', color=[colors[x] for x in trips_percent.index])
plt.xlabel('User Type')
plt.ylabel('Percentage in %')
plt.title('Annual average percentage of use of user types')
plt.ylim(0,100)
plt.yticks(range(0, 110, 10))
plt.show()
```

The bar chart illustrates the average annual usage of the user types. It is clearly visible that subscribers make about 80% of all trips.
